```python
# Create the pandas DataFrame with column name is provided explicitly
df_pixel = pd.DataFrame(columns=['time','loc','x', 'r','g','b','unknown'])

for p in range(0,2881,1):
    for f in range(4,5,1):
        filename = '24.11.14 10 mM Bicarbonate, 20 mM HCl with Droplets_Top Slide_R_p%02d_0_A01f%02dd4.PNG'%(p,f)
        print(directory+filename)
        try:
            image = pims.open(directory+filename)[0]
            #if len(image.shape) > 2 and image.shape[2]==4:
                #image = image[:,:,:3]
            img = image
            a = image.shape
            print('first =', a)
            #Lx = int(img.shape[1] / 10)
            rgb = np.mean(img,axis=0)
            b = rgb.shape
            print('second=', b)
            #rgb = np.mean(np.reshape(rgb,(Lx,10,3)), axis=1)
            #c = rgb.shape
            #print('third =', c)
            rows = {'time': p*np.ones((Lx,), dtype=int),
                    'loc': f*np.ones((Lx,), dtype=int),
                    'x': np.arange(Lx),
                    'r': rgb[:,0],
                    'g': rgb[:,1],
                    'b': rgb[:,2]}
                   #'unknown': rgb[:,3]}
            new_df_pixel = pd.DataFrame(rows)
            df_pixel = pd.concat([df_pixel,new_df_pixel], ignore_index=True)
        except:
            print('no such file!')
```

```python
loc = 4 # time index
time = 1 # space index

idx2 = (df_pixel['time']==time) & (df_pixel['loc']==loc)



plt.plot(df_pixel['x'][idx2], df_pixel['r'][idx2], 'r', label='R')
plt.plot(df_pixel['x'][idx2], df_pixel['g'][idx2], 'g', label='G')
plt.plot(df_pixel['x'][idx2], df_pixel['b'][idx2], 'b', label='B')
plt.xlabel('$x$ (pixels)')
plt.ylim(50,120)
plt.legend(frameon=False);
```

```python
df_pixel.head()
```

```python
# add an empty columns
df_pixel['X'] = np.nan
df_pixel['Y'] = np.nan
df_pixel['Z'] = np.nan
df_pixel['x_color'] = np.nan
df_pixel['y_color'] = np.nan

for index, row in df_pixel.iterrows():
    rgb = sRGBColor(row['r']/255, row['g']/255, row['b']/255)
    xyz = convert_color(rgb, XYZColor, target_illuminant='d50')
    df_pixel.at[index, 'X'] = xyz.xyz_x
    df_pixel.at[index, 'Y'] = xyz.xyz_y
    df_pixel.at[index, 'Z'] = xyz.xyz_z
    S = xyz.xyz_x + xyz.xyz_y + xyz.xyz_z
    df_pixel.at[index, 'x_color'] = xyz.xyz_x / S
    df_pixel.at[index, 'y_color'] = xyz.xyz_y / S
```
